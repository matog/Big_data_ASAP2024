## Paso 1: Librerias y dataset

```python
# Importamos pandas para trabajar con DF
import pandas as pd
```

```python
# Importamos Matplotlib para poder configurar los gráficos
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Importamos Seaborn para potenciar matplotlib con visualizaciones atractivas
import seaborn as sns
```

### Carga del Dataset

```python
# Import the Pokemon dataset and set index position to column 0, or the first column
df = pd.read_csv('https://raw.githubusercontent.com/matog/Flacso_ciencia_de_datos_python_2022/main/Clase2/data/pokemon.csv', sep = ';', index_col = 0)
```

```python
# Leemos los primeros 5 registros
df.head()
```

```python
#Descripción del dataset
df.describe()
```

## Paso 2: Utilizamos las funciones gráficas de Seaborn

```python
# Default scatterplot
sns.lmplot(x='Attack', y='Defense', data=df);
```

```python
# Customizamos el scatter plot para eliminar la linea de regresión y ajustar los colores por stage
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=False, hue='Stage')
```

## Paso 3: Utilizamos Matplotlib para customizar los ejes

```python
# Graficamos con Seaborn
sns.lmplot(x='Attack', y='Defense', data=df, fit_reg=False, hue='Stage')

# Tweak de los ejes x e y con Matplotlib
plt.ylim(0, None)
plt.xlim(0, None)
```

## Paso 4: Pandas con Seaborn

```python
# Generamos un boxplot
sns.boxplot(data=df);
```

```python
# Borramos columnas innecesarias
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
```

```python
# Graficamos nuevamente
sns.boxplot(data=stats_df);
```

## Paso 5: Ajustamos el theme de Seaborn


Listado de themes de seaborn: [Link](https://www.python-graph-gallery.com/104-seaborn-themes)

```python
# Cambiamos el theme
sns.set_style('white')

# Armamos un violin plot para ver las distribuciones
sns.violinplot(x='Type 1', y='Attack', data=df);
```

## Paso 6:Creamos una paleta de colores con Python Lists

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
# Use palette argument to pass in custom colors
sns.violinplot(x='Type 1', y='Attack', data=df, palette=pkmn_type_colors);
```

```python
# Creamos un gráfico Swarm para mostrar puntos individuales
sns.swarmplot(x='Type 1', y='Attack', data=df, palette=pkmn_type_colors)
```

## Paso 7: Gráficos superpuestos

```python
# Adjustamos el tamaño de la figura utilizando la función figure () y el atributo figsize en Matplotlib
plt.figure(figsize=(20,12))
```

```python
# Superponemos graficos violin y swarn
sns.violinplot(x='Type 1', y='Attack', data=df, inner=None, palette=pkmn_type_colors)
sns.swarmplot(x='Type 1', y='Attack', data=df, color='g', alpha=0.7)

# Generamos el título
plt.title('Tipo de ataque');
```

## Paso 8: Uniendo todo

```python
stats_df
```

```python
# Convertimos las 6 variables en una sóla, creando nuevas filas para cada una.
melted_df = pd.melt(stats_df, 
                    id_vars = ["Name", "Type 1", "Type 2"], # Variables to keep
                    var_name = "Stat") # Name of melted variable


melted_df.tail(15)
# melted_df['Stat'].unique()
```

```python
melted_df.loc[melted_df['Name']=='Nidorina']
```

```python
# Comparamos las dimensiones del DF para ver si realmente se efectuó el _melt_
print (stats_df.shape)
print (melted_df.shape)
```

```python
# Ajustamos el tamaño del gráfico utilizando la función figure() y el atributo figsize en Matplotlib
plt.figure(figsize=(20,12))

# Creamos un swarmplot con los datos _melted_
sns.swarmplot(x='Stat', y='value', data=melted_df, hue='Type 1')
```

```python
# Aumentamos el tamaño de la figura
plt.figure(figsize=(20,12))

# Separamos los puntos por color
sns.swarmplot(x='Stat', y='value', data=melted_df, hue='Type 1', split=True, palette= pkmn_type_colors)

# Alternativa con Dogde en lugar de Split y con menor tamaño en los puntos (s)
#sns.swarmplot(x='Stat', y='value', data=melted_df, hue='Type 1', dodge=True, palette= pkmn_type_colors, s = 5)

# Cambios el eje Y
plt.ylim(0,260)

# Re-posicionamos la leyenda en el lado derecho de la figura
plt.legend(bbox_to_anchor=(1,1), loc=2);
```

## Paso 9: Ejemplos de visualizaciones

### 9.1 - Heatmaps -> Ayuda para visualizar datos en forma de matriz

```python
# Generamos la matriz de correlaciones con el método corr()
plt.figure(figsize=(20,12))
corr = stats_df.corr()

# Generamos el heatmap basado en las correlaciones
sns.heatmap(corr)
```

```python
corr
```

### 9.2 - Gráficos de distribuciones (a.ka. Histogramas)  -> Ayuda a visualizar la distribución de la información numérica

```python
# Generamos el gráfico de la distribución con el método distplot()
plt.figure(figsize=(20,12))
sns.distplot(df['Speed'])
```

### 9.3 - Bar Plots (a.k.a Count Plots) -> Ayuda a visualizacar datos categóricos

```python
# Creamos un Bar Plot
plt.figure(figsize=(20,12))
sns.countplot(x='Type 1', data=df)

# Rotamos las etiquetas el eje x para que se lea mejor
plt.xticks(rotation=90);
```

### 9.4 - Factor Plot -> Sirve para separar gráficos con información categórica

```python
# Creamos un Factor Plot
plt.figure(figsize=(20,12))
g = sns.factorplot(x='Type 1', y='Attack', data=df, hue='Stage', col='Stage', kind='swarm')

# Rotamos la etiqueta del eje x para todos los gráficos
g.set_xticklabels(rotation= -45)
```

### 9.5 - Density Plot -> Ayuda a mostrar la distribución con dos variables

```python
# Density Plot
sns.kdeplot(df.Attack, df.Defense)
```

### 9.6 - Joint Distribution Plot ->  Combina un scatter y un gráfico de distribución para poder sacar insights sobre datos bi variados

```python
# Joint Distribution Plot
sns.jointplot(x='Attack', y='Defense', data=df)
```
